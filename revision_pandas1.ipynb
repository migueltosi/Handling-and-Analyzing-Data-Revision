### How to create new dataframes based on Python data structures (list, dictionaries or tuples)

```python
data = [('Azul', 'Amarelo', 'Branco'), ('Verde', 'Rosa', 'Preto'), ('Roxo', 'Laranja', 'Vermelho')]
index = ['familia_macaco' + str(i+1) for i in range (3)]
columns = ['Mãe', 'Pai', 'Filho']
macacos = pd.DataFrame(data = data, columns = columns, index = index)
macacos
```

```python
data = [('Preto', 'Cinza', 'Lilás'), ('Verde', 'Verde', 'Verde'), ('Amarelo', 'Roxo', 'Preto')]
index = ['familia_elefante' + str(i+1) for i in range (3)]
columns = ['Mãe', 'Pai', 'Filho']
elefantes = pd.DataFrame(data = data, columns = columns, index = index)
elefantes
```

```python
data = [('Branco', 'Branco', 'Branco'), ('Branco', 'Branco', 'Preto'), ('Branco', 'Branco', 'Branco')]
index = ['familia_ovelha' + str(i+1) for i in range (3)]
columns = ['Mãe', 'Pai', 'Filho']
ovelhas = pd.DataFrame(data = data, columns = columns, index = index)
ovelhas
```

### How to concatenate dataframes (remembering the axis)

```python
animais = pd.concat([macacos, elefantes, ovelhas])
animais
```

## Frequency

### Selection forms and frequencies

```python
cores_arco_iris = ['Vermelho', 'Laranja', 'Amarelo', 'Verde', 'Azul', 'Roxo']
cai = cores_arco_iris
```

#### Selection with OR (|) condition

```python
coluna = 'Mãe'
maes_arco_iris = (animais[coluna] == cai[0]) | (animais[coluna] == cai[1]) | (animais[coluna] == cai[2]) | (animais[coluna] == cai[3]) | (animais[coluna] == cai[4]) | (animais[coluna] == cai[5])
```

```python
coluna = 'Pai'
pais_arco_iris = (animais[coluna] == cai[0]) | (animais[coluna] == cai[1]) | (animais[coluna] == cai[2]) | (animais[coluna] == cai[3]) | (animais[coluna] == cai[4]) | (animais[coluna] == cai[5])
```

```python
coluna = 'Filho'
filhos_arco_iris = (animais[coluna] == cai[0]) | (animais[coluna] == cai[1]) | (animais[coluna] == cai[2]) | (animais[coluna] == cai[3]) | (animais[coluna] == cai[4]) | (animais[coluna] == cai[5])
```

#### Selection with the AND (&) condition

```python
familia_arco_iris = maes_arco_iris & pais_arco_iris & filhos_arco_iris
animais[familia_arco_iris]
```

## Handling Missing Data

### How to identify null values ​​(missing values)

```python
familia_animal = pd.concat([macacos, elefantes, ovelhas], axis = 1)
familia_animal.columns = ['Mãe', 'Pai', 'Filho', 'Vó1', 'Vô1', 'Vó2', 'Vô2', 'Tio', 'Tia']
familia_animal
```

#### The isnull() method indicates whether values ​​are null

```python
familia_animal.isnull()
```

#### The notnull() method returns the opposite of the isnull() method

```python
familia_animal.notnull()
```

#### The info() method is also a way to check for missing values

```python
familia_animal.info()
```

### How to remove null values ​​with dropna() method

```python
familia_animal.dropna()
```

### How to replace missing values ​​with fillna() method

```python
familia_animal.fillna('Transparente')
```

### Interpolation methods: ffill, bfill and mean()

```python
familia_animal.fillna(method = 'ffill') #de cima para baixo, copia o anterior
```

```python
familia_animal.fillna(method = 'bfill') #de baixo para cima, copia o anterior
```

```python
#fazendo os dois anteriores juntos

familia_animal = familia_animal.fillna(method = 'ffill')
familia_animal = familia_animal.fillna(method = 'bfill')
familia_animal
```
